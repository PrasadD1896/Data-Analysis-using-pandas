#### Series:

It is a one dimensional entity in pandas.

```python
dict = {'a':1,'b':2,'c':3}

series1 = pd.Series(dict)
print(series1)
```

```python
lst = [1,2,3,4,5]

series2 = pd.Series(lst)
print(series2)
```

#### Dataframe:

It is a 2 dimensional entity in pandas.

```python
dict = {'Name':['Rohit','Anuj','Vidyesh'],
       'Age': [28,27,28],
       'City': ['Pune','Dhule','Nashik']}

df1 = pd.DataFrame(dict)
print(df1)
```

```python
lst = [['Rohit',28,'Pune'],
      ['Anuj',27,'Dhule'],
      ['Vidyesh',28,'Nashik']]

df2 = pd.DataFrame(lst,columns = ['Name','Age','City'])
print(df2)
```
